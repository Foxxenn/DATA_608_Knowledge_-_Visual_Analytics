# Data Visualization

We use graphs to visualize where the data lies within our ranges. From the visualization, we observe that Mississippi (MS) has the highest rate, 29.1 in 2021, while Massachusetts (MA) has the lowest rate, 3.4 in the same year. They are both highlighted by the DARK BLUE and the YELLOW colors. To enhance clarity, we need to order the rates. One approach could be to aggregate the rates from all years, divide the total by 9 (the number of years we have), and then sort the resulting values.

```python
heatmap_data = mortality_data.pivot("STATE", "YEAR", "RATE")
```

```python
plt.figure(figsize=(24,15))  
sns.heatmap(heatmap_data, annot=True, fmt=".1f", cmap="YlGnBu")
plt.title("Death Rates by State 2005-2021", fontsize=30)
plt.xticks(fontsize=30)  
plt.yticks(fontsize=18)
plt.xlabel("Year", fontsize=30) 
plt.ylabel("State", fontsize=20) 
plt.yticks(rotation=0) 

plt.show()
```

Here is the data aggregated by the rates of the added years.

```python
aggregated_data = mortality_data.groupby('STATE')['RATE'].sum().reset_index()

sorted_aggregated_data = aggregated_data.sort_values(by='RATE', ascending=False)


sorted_aggregated_data['RATE'] = round((sorted_aggregated_data['RATE']/ 9), 2)

sorted_aggregated_data.head()
```

I pivoted the data to facilitate its visualization in the heatmap

```python
heatmap_data = sorted_aggregated_data.set_index("STATE")
heatmap_data.head()
```

We plotted the information on a graph, clearly showing that Mississippi leads in gun rates with 22.8, while the lowest rate is in Massachusetts at 3.4. Remember to divide these rates by 9 years. Surprisingly, New York and California have very low rates. It appears that Republican-owned states generally have worse gun problems, while Democratic states tend to have much lower gun mortality rates

```python
plt.figure(figsize=(24,15))  
sns.heatmap(heatmap_data, annot=True, fmt=".1f", cmap='YlGnBu')
plt.title("Death Rates by State", fontsize=30)
plt.xticks(fontsize=30)  
plt.yticks(fontsize=18)
# plt.xlabel("RATE", fontsize=30) 
plt.ylabel("STATE", fontsize=20) 
plt.yticks(rotation=0) 

plt.show()
```

# Conclusion

The question of whether gun laws reduce mortality rates is complex. In many cases, the answer is yesstates with gun laws often experience a decrease in gun-related deaths. For instance, New York (NY) and California (CA), which have some of the strictest gun laws, generally exhibit lower mortality rates. However, some states see an increase in mortality rates despite having strict gun laws, as seen in Illinois (IL).
